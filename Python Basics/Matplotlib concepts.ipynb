```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
a = np.arange(0,10,2)
```

```python
plt.plot(x,y,'g--')
plt.xlabel('Xlabel')
plt.ylabel('ylabel')
plt.title('My first plot')
plt.show()
```

**Creating Subplots**

```python
nrows = 1
ncols = 2
plt.subplot(nrows , ncols , 1)
plt.plot(x,y,'r--')
plt.title('First plot')
plt.subplot(nrows , ncols , 2)
plt.plot(y,x,'g')
plt.title('Second plot')

#This will create multiple plots on the same canvas
```

**Using object oriented method to instantiate figures/plots**

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])  #Left,Bottom,Width,Height ranges from 0 to 1

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
#Using object oriented approach to plot two figures

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,1,1])
axes1.plot(x,y,'r+')
axes1.set_title('Larger Plot')
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(x,y,'b--')
axes2.set_title('Smaller Plot')
```

**Creating subplots using object oriented method**

```python
#here we can iterate through axes and create diff plots
fig , axes = plt.subplots(nrows = 3, ncols = 3)
plt.tight_layout()
```
